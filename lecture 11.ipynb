import matplotlib.pyplot as plt
import pandas as pd

class CityScatterPlot:
    def __init__(self, data):
        """Khởi tạo dữ liệu"""
        self.cities = pd.DataFrame(data)
        self.fig, self.ax = plt.subplots()

    def create_plot(self):
        """Tạo scatter plot theo tọa độ latitude và longitude"""
        sc = self.ax.scatter(
            self.cities["longd"], 
            self.cities["latd"], 
            s=self.cities["population_total"] / 500, 
            c=self.cities["area_land_km2"], 
            cmap='coolwarm', 
            alpha=0.7, 
            edgecolors='k'
        )

        # Thêm nhãn
        self.ax.set_xlabel("Longitude")
        self.ax.set_ylabel("Latitude")
        self.ax.set_title("Scatter Plot of Cities by Latitude and Longitude\n(Size: Population, Color: Land Area)")

        # Thêm color bar
        cbar = self.fig.colorbar(sc)
        cbar.set_label("Land Area (km²)")

    def show_plot(self):
        """Hiển thị biểu đồ"""
        plt.show()

Điểm mạnh của cách tiếp cận OOP này:

```python
fig, axes= plt.subplots(nrows=2, ncols=3, figsize =(6,4),sharex=True, sharey=True)
for i in range(2):
    for j in range (3):
        axes[i,j].text(0.4,0.4,str((i,j)))


# Vẽ 6 biểu đồ scatter plot vào các subplot
fig, axes = plt.subplots(2, 3, figsize=(12, 8), sharex=True, sharey=True)

# Vẽ biểu đồ
for ax, plot_type, (index, row) in zip(axes.flat, cities.iterrows()):
    if plot_type == 'scatter':
        ax.scatter(row["longd"], row["latd"], s=row["population_total"] / 500, c='red', alpha=0.7)
    elif plot_type == 'bar':
        ax.bar(["Area Land"], [row["area_land_km2"]], color='blue')
    elif plot_type == 'hist':
        ax.hist(cities["population_total"], bins=5, color='green', alpha=0.7)
    elif plot_type == 'line':
        ax.plot(cities["latd"], cities["population_total"], marker='o', linestyle='-', color='purple')
    elif plot_type == 'box':
        ax.boxplot(cities["population_total"], vert=False)
    elif plot_type == 'pie':
        ax.pie([row["population_total"], sum(cities["population_total"]) - row["population_total"]], labels=[row["city"], "Others"], autopct='%1.1f%%', colors=['orange', 'gray'])
    
    ax.set_title(f"{plot_type.capitalize()} Plot")



# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
