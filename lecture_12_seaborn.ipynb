```python
import matplotlib.pyplot as plt 
```

```python
import seaborn as sns
import pandas as pd
url1= 'https://raw.githubusercontent.com/ntlong81/Finacial_bigdata_analysis_BFT64/main/Data/Pokemon.csv'
df=pd.read_csv(url1)

```

```python
# Load the example flights dataset
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot(index = "month",
                              columns="year",
                              values ="passengers")

flights.head()
```

```python
# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d",
            linewidths=1.5);
```

## diamonds

```python
diamonds_df = sns.load_dataset('diamonds')
diamonds_df.head()
```

```python
columm_drop = ['cut','clarity','color']
diamonds_new = diamonds_df.drop(columm_drop, axis=1)
diamonds_new.head()
```

```python
diamonds_corr= diamonds_new.corr()
diamonds_corr 
```

```python
 
import numpy as np
f, ax = plt.subplots(figsize=(9, 6))
mask = np.triu(np.ones_like(diamonds_corr, dtype=bool))

sns.heatmap(diamonds_corr, mask= mask, annot=True, fmt=".2f", cmap=['lightgreen','lightblue'], linewidths=0.5) 
```
