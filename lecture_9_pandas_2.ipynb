# concat()

```python
import pandas as pd
df1 = pd.DataFrame([['a', 1], ['b', 2]],columns=['letter', 'number'])
df2 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']],columns=['letter', 'number', 'animal'])
```

```python
print(df1)
print(df2)
```

```python
pd.concat([df1,df2], axis=1,join='outer')
```

```python
pd.concat([df1,df2], axis=1,join='inner')
```

```python
pd.concat([df1,df2], axis=0,join='outer')
```

```python
pd.concat([df1,df2], axis=0,join='inner')
```

```python
import pandas as pd
url1 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv'
url2 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv'
url3 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv'

pop = pd.read_csv(url1)
areas = pd.read_csv(url2)
abbrevs = pd.read_csv(url3)
```

```python
pop.head()
import pandas as pd
```

```python
areas.head()
```

```python
abbrevs.head()
```

```python
new= pd.merge(abbrevs,areas)
df=pd.merge(new,pop,right_on='state/region',left_on="abbreviation")
df
```

```python
df.isnull().sum()
```

```python
df_total = df[df['ages']=='total']
df_totall=df_total[df_total['year']==2010]
df_totall
```

```python
density = (df_totall['population'])/(df_totall['area (sq. mi)'])
df_totall['density']=density
df_totall
```

```python
df_totall.sort_values(by='density',ascending=False,inplace=False)
```

# filtering and transformation


```python
import pandas as pd
data = {'key': ['A', 'B', 'A', 'B', 'A', 'C'],
        'data1': [1, 2, 3, 4, 5, 6],
        'data2': [7, 8, 9, 10, 11, 12]}

df = pd.DataFrame(data)
print(df.groupby('key').sum())
df.groupby('key').filter(lambda x: x.data1.sum()>=5)
```

```python
df.groupby('key').transform(lambda x: x-x.mean())
```

```python
def norm_by_data2(x):
    # x is a dataframe of group value
    x['data1']/=x['data2'].sum()
    return x
df.groupby('key').apply(norm_by_data2)
```

```python
# task2:
pd.pivot_table(data=titanic, index='sex',columns=['pclass','embark_town'], values='fare')
```

```python
# task3:
pd.pivot_table(data=titanic, index='deck',columns='pclass', values='survived', aggfunc='sum')
```

```python
# task 4:
pd.pivot_table(data=titanic, index='class',columns='sex', values='age', aggfunc=['median','mean'])
```

```python
# task 6:
age_group = pd.cut(titanic['age'],
                   bins=[0,12,18,35,60,80], 
                   labels = ['child','teen','sdult','midle age','senior'])
# chia muc tuoi tu 0 den 80 trong do 80 la so tuoi lớn nhất của hành khchs trên tàu va gan lableslables 
# thêm age group và dataframe
titanic['age_group']=age_group
pd.pivot_table(data=titanic, index='sex',columns='age_group', values='survived')
```
